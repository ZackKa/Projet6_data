```python
# ===============================
# ETAPE 1 : NETTOYAGE ET FILTRAGE
# ===============================

# 1️⃣ Suppression des colonnes peu utiles
# Ces colonnes n'apportent pas d'information utile pour l'analyse ou le modèle
columns_to_drop = ["Comments", "Outlier", "YearsENERGYSTARCertified"]
building_consumption_clean = building_consumption.drop(columns=columns_to_drop)


# 2️⃣ Correction des valeurs incohérentes de LargestPropertyUseTypeGFA
# -> Si LargestPropertyUseTypeGFA > PropertyGFATotal, alors la donnée est impossible.
# On va donc remplacer ces valeurs par la MÉDIANE calculée pour chaque type de propriété.

# Création d'un masque qui repère les valeurs incohérentes
mask_incoherent_gfa = (
    building_consumption_clean["LargestPropertyUseTypeGFA"] 
    > building_consumption_clean["PropertyGFATotal"]
)

# BOUCLE ⬇⬇⬇ : pour chaque type de bâtiment, on calcule une médiane et on remplace
for prop_type in building_consumption_clean["PrimaryPropertyType"].unique():
    # On calcule la médiane DES VALEURS COHÉRENTES pour ce type de bâtiment
    median_gfa = building_consumption_clean.loc[
        (building_consumption_clean["PrimaryPropertyType"] == prop_type) &
        (~mask_incoherent_gfa),  # valeurs cohérentes seulement
        "LargestPropertyUseTypeGFA"
    ].median()
    
    # On remplace les valeurs incohérentes par la médiane correspondante
    building_consumption_clean.loc[
        (building_consumption_clean["PrimaryPropertyType"] == prop_type) &
        mask_incoherent_gfa,  # valeurs incohérentes seulement
        "LargestPropertyUseTypeGFA"
    ] = median_gfa


# 3️⃣ Suppression des lignes incohérentes dans d'autres colonnes
# On garde UNIQUEMENT les lignes avec des valeurs strictement positives
building_consumption_clean = building_consumption_clean[
    (building_consumption_clean["SiteEnergyUse(kBtu)"] > 0) &    # énergie positive
    (building_consumption_clean["NumberofFloors"] > 0) &         # ≥ 1 étage
    (building_consumption_clean["NumberofBuildings"] > 0)        # ≥ 1 bâtiment
]


# 4️⃣ Filtrage : on garde uniquement les bâtiments NON résidentiels
non_residential_types = [
    "NonResidential",
    "Nonresidential COS",
    "Nonresidential WA",
    "SPS-District K-12",
    "Campus"
]

building_consumption_clean = building_consumption_clean[
    building_consumption_clean["BuildingType"].isin(non_residential_types)
]


# 5️⃣ Affichage de contrôle
# Affiche le nombre de lignes et colonnes restants
print("Shape après nettoyage et filtrage :", building_consumption_clean.shape)

# Affiche la répartition des types de bâtiment restants
print("\nRépartition des BuildingType après filtrage :")
print(building_consumption_clean["BuildingType"].value_counts())
```
